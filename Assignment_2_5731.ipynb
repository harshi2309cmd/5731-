```python
from bs4 import BeautifulSoup as bs
import requests
link = 'https://www.imdb.com/title/tt6751668/reviews?ref_=tt_urv'
page = requests.get(link)
```

```python
import urllib.request
req = urllib.request.Request(url=link)
with urllib.request.urlopen(req) as f:
 print(f.read().decode('utf-8'))
```

```python
page
```

```python
soup = bs(page,'html.parser')
```

```python
print(soup.prettify())
```

```python

titles=soup.find_all(class_='title')
titles
```

```python
review_title=[]
for i in range(0,len(titles)):
    review_title.append(titles[i].get_text())


review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
names = soup.find_all('span',class_='display-name-link')
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
time = soup.find_all('span',class_='review-date')
time
```

```python
review_time = []
for i in range(0,len(time)):
    review_time.append(time[i].get_text())
review_time
```

```python
rating = soup.find_all('span',class_='rating-other-user-rating')
rating
```

```python
star_rating = []
for i in range(0,len(rating)):
    star_rating.append(rating[i].get_text())
star_rating
star_rating[:] = [rating.lstrip('\n') for rating in star_rating]
star_rating
star_rating[:] = [rating.rstrip('\n') for rating in star_rating]
star_rating

star_rating.append('5/10')
star_rating
```

```python
review = soup.find_all(class_='text show-more__control')
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content

review_content[:] = [review.lstrip('\n') for review in review_content]
review_content
review_content[:] = [review.rstrip('\n') for review in review_content]
review_content
```

```python
cust_name
review_content
star_rating
review_time
review_title
```

```python
import pandas as pd
df = pd.DataFrame()
df['Customer Name']=cust_name
df['Review Name']=review_title
df['Time']=review_time
df['Content']=review_content
df['Rating']=star_rating
df
```

```python
df.to_csv(r'E:\reviews.csv',index=True)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

Review_title=[]
author_list=[]
venue_list=[]
pubyear_list=[]
abstract_list=[]
x = 0
while x < 1000:
  headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}
  url = 'https://citeseerx.ist.psu.edu/search;jsessionid=BC906AC2234F0229BE97EBD715D4F2EB?q=test&t=doc&sort=rlv&start=x'
  response=requests.get(url,headers=headers)
  soup=BeautifulSoup(response.content,'lxml')
  user_review=soup.find_all('div',class_='result')
  for container in user_review:
    titles=container.a.text
    Review_title.append(titles)
    author=container.find('span', class_='authors').text
    author_list.append(author)
    venue=container.find('span', class_='pubvenue')
    venue_list.append(venue)
    year=container.find('span', class_='pubyear').text
    pubyear_list.append(year)
    abstract=container.find('div',class_='pubabstract')
    abstract_list.append(abstract)
  x = x+10

dict = {'Title': Review_title, 'author': author_list, 'year': pubyear_list,'venue': venue_list,'abstract': abstract_list} 
    
df = pd.DataFrame(dict)


pd.set_option("display.max_rows", None, "display.max_columns", None,'display.max_colwidth', -1)

df
```
