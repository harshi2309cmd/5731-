```python
#importing required libraries
import pandas as pd
train = pd.read_csv ('assignment.txt', sep='delimiter', header=None)
train.columns =['Sent']
data_top = train
# display 
data_top .head()
```

```python
train['word_count'] = train['Sent'].apply(lambda x: len(str(x).split(" ")))

#text with word count
print('\n\nWORD COUNT\n\n')
train[['Sent','word_count']].head()
```

```python

# number of characters
print('\n\nNUMBER OF CHARACTERS\n\n')
train['char_count'] = train['Sent'].str.len() ## this also includes spaces
train[['Sent','char_count']].head()
```

```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))
print('\n\nAverage Word Length\n\n')
train['avg_word'] = train['Sent'].apply(lambda x: avg_word(x))
train[['Sent','avg_word']].head()
```

```python

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = stopwords.words('english')

train['stopwords'] = train['Sent'].apply(lambda x: len([x for x in x.split() if x in stop]))
train[['Sent','stopwords']].head()
```

```python
train['hastags'] = train['Sent'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
train[['Sent','hastags']].head()
```

```python
train['numerics'] = train['Sent'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
train[['Sent','numerics']].head(10)
```

```python
train['upper'] = train['Sent'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
train[['Sent','upper']].head()
```

```python
train['Sent'] = train['Sent'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['Sent'].head()
```

```python
train['Sent'] = train['Sent'].str.replace('[^\w\s]','')
train['Sent'].head()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['Sent'] = train['Sent'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
train['Sent'].head()
```

```python

freq = pd.Series(' '.join(train['Sent']).split()).value_counts()[:10]
freq
```

```python
freq = list(freq.index)
train['Sent'] = train['Sent'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
train['Sent'].head()
```

```python
freq = pd.Series(' '.join(train['Sent']).split()).value_counts()[-10:]
freq
```

```python
freq = list(freq.index)
train['Sent'] = train['Sent'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
train['Sent'].head()
```

```python
#term frequency
tf1 = (train['Sent'][0:10]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
#pos speech
from nltk.tokenize import word_tokenize
import nltk
nltk.download('averaged_perceptron_tagger')
text_tags = []
for s in train['Sent']:
  text = word_tokenize(str(s))
  text_tags.append(nltk.pos_tag(text))
text_tags
```

```python

import spacy
# nlp = spacy.load("en_core_web_sm")
import en_core_web_sm
nlp = en_core_web_sm.load()

entity_types = ['PERSON','NORP','FAC','ORG','GPE','LOC','PRODUCT','EVENT','WORK_OF_ART','LAW',
                'LANGUAGE','DATE','TIME','PERCENT','MONEY','QUANTITY','ORDINAL','CARDINAL']

entity_name = ['person','Nationality','Building','Institution','country','location','PRODUCT','EVENT','Title','LAW',
                'LANGUAGE','DATE','TIME','PERCENT','MONEY','QUANTITY','ORDINAL','CARDINAL']

def entity_extraction(sentence):
    entity_list = []
    doc = nlp(sentence)
    for ent in doc.ents:
        entity_list.append([ent.text, ent.start_char, ent.end_char, ent.label_])
    
    return entity_list
```

```python
#Inverse Document Frequency
import numpy as np

for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(train.shape[0]/(len(train[train['Sent'].str.contains(word)])))

tf1
```

```python
#Term Frequency – Inverse Document Frequency (TF-IDF)
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```
