```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))
print('\n\nAverage Word Length\n\n')
train['avg_word'] = train['Sent'].apply(lambda x: avg_word(x))
train[['Sent','avg_word']].head()
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('averaged_perceptron_tagger')
text_tags = []
for s in train['Sent']:
  text = word_tokenize(str(s))
  text_tags.append(nltk.pos_tag(text))
text_tags
```

```python
nouns=0
verbs=0
adverbs=0
adjectives=0
for i in text_tags:
  for j in i:
    if 'NN' in j:
      nouns+=1
    if 'VB' in j:
      verbs+=1
    if 'ADJ' in j:
      adjectives+=1
    if 'ADV' in j:
      adverbs+=1
print("nouns:",nouns)  
print("verbs:",verbs) 
print("adverbs:",adverbs) 
print("adjectives:",adjectives)  
```

```python
!pip install benepar
import benepar
import nltk
benepar.download('benepar_en3')
bane_parser = benepar.Parser("benepar_en3")
for line in train['Sent']:
  try:
    tree = bane_parser.parse(line)
    print(tree)
  except:
    print("No Parse Tree")
    continue
```

```python
import en_core_web_sm
nlp = en_core_web_sm.load()
list=[]
for s in train['Sent']:
  doc = nlp(s)
  for X in doc.ents:
    if X.text and X.label_:
      entity=[(X.text, X.label_)]
      list.append(entity)
print(list)

CARDINAL=0
ORG=0
PERSON=0
DATE=0
MONEY=0
QUANTITY=0
ORDINAL=0
GPE=0

for i in list:
  for j in i:
    if 'CARDINAL' in j:
      CARDINAL+=1
    if 'ORG' in j:
      ORG+=1
    if 'PERSON' in j:
      PERSON+=1
    if 'DATE' in j:
      DATE+=1
    if 'MONEY' in j:
      MONEY+=1
    if 'QUANTITY' in j:
      QUANTITY+=1
    if 'ORDINAL' in j:
      ORDINAL+=1
    if 'GPE' in j:
      GPE+=1

print("CARDINAL:",CARDINAL) 
print("ORG:",ORG) 
print("PERSON:",PERSON) 
print("DATE:",DATE) 
print("MONEY:",MONEY) 
print("QUANTITY:",QUANTITY) 
print("ORDINAL:",ORDINAL) 
print("GPE:",GPE) 
```

**Parsing**

parsing is meant to the task of creating a parse tree from a given sentences.

**Constituency Parsing**

it is known for context-free grammars and it is divided into various constituents where the sub-phrases belong to specific category.The grammar provides us to bulid various valid sentences and also to set various rules and also to extract the syntactical structure of a given sentence according to the grammar such nouns,verbs,adverbs and adjectives.

**Dependency Parsing**

Dependency parsing doesnt make use of phrasal constituents or sub-phrases. Instead of that it uses the syntax of the sentence which is  expressed in terms of dependencies between words 
it is directed to edges between words in a graph.

```python
# function to remove leading zeros
def removeZeros(ip):
    new_ip = ".".join([str(int(i)) for i in ip.split(".")])  
    return new_ip 
```
