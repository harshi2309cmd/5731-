```python
# Write a Python program to count the number of characters in a string.

def count(str):
    dict = {}
    for i in str:
        keys = dict.keys()
        if i in keys:
            dict[i] += 1
        else:
            dict[i] = 1
    return dict
print(count('google.com'))
```

```python
#Write a Python program to sum all the items in a list.

def sum(items):
    n = 0
    for x in items:
        n += x
    return n
print(sum([1,-2,9]))
```

```python
 #Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '\$', except the first char itself.
 
 def occur(str):
  char = str[0]
  str = str.replace(char, '$')
  str = char + str[1:]

  return str

print(occur('restart'))
```

```python
#Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. 
def Duplicates(array):
    result = []
    for x in array:
        if (array.count(x)>1) and (x not in result):
            result.append(x)
    if len(result)==0:
        return -1
    else:
        return result
print(Duplicates([3,2,4,3,2,5,3,7]))
print(Duplicates([4,2,3,1,5]))
```

```python
#Write a Python program to select all the Sundays of a specified year. 
import pandas as pd
def allsundays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-SUN').strftime('%m/%d/%Y').tolist()

allsundays(2021)
```
