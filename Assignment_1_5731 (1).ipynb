```python
import random
numList=[]
for x in range(12):
  integer = random.randint(1,100)
  numList.append(integer)
print("Randomly generated 12 numbers between 1 to 100: ")
for i in range(len(numList)):
  print(numList[i], end=' ',)

numList.sort()
print("\nAscending Order of the 12 randomly generated numbers: ")
for i in range(len(numList)):
  print(numList[i], end=' ',)

print("\nDescending Order of the 12 randomly generated numbers: ")
numList.sort(reverse=True)
for i in range(len(numList)):
  print(numList[i], end=' ',)

```

```python
import re
def commonWords(A,B):
   result=[]
   #split the sentences
   s1=A.split()
   s2=B.split()
   result.append(list(set(s1)&set(s2)))
   #remove punctuation 
   emptyString=""
   n1=re.sub(r'[^\w\s]',emptyString,A)
   n2=re.sub(r'[^\w\s]',emptyString,B)
   print('string 1 without punctuation:',n1)
   print('string 2 without punctuation:',n2)
   #check vowels and number of times it is repeated
   vowels = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}
   for letter in A:
    if letter.lower() in vowels:
     vowels[letter.lower()] += 1
     print("\nVowel frequency ",vowels)

   return result
```

```python
commonWords("python is good and dynamic programming language!","I am good , quick and dynamic programmer")
```

```python
#2

unique_year = sorted(list(data['year'].unique()))
occupied_houses = data[data['title_field'] == 'Occupied Housing Units'].sort_values(by=['year'])
vacant_houses = data[data['title_field'] == 'Vacant Housing Units'].sort_values(by=['year'])
new_house_df = pd.DataFrame()
new_house_df['year'] = unique_year
new_house_df['Vacant House'] = vacant_houses['value'].values
new_house_df['Occupied House'] = occupied_houses['value'].values
new_house_df['Percent of Vacant House'] = [(new_house_df['Vacant House'][loop1] * 100) /\
                                           (new_house_df['Vacant House'][loop1] + new_house_df['Occupied House'][loop1])\
                                           for loop1 in range(len(new_house_df))]
new_house_df['Percent of Occupied House'] = [(new_house_df['Occupied House'][loop1] * 100) /\
                                           (new_house_df['Vacant House'][loop1] + new_house_df['Occupied House'][loop1])\
                                           for loop1 in range(len(new_house_df))]
new_house_df
```

```python
#3

max_vacant_houses = max(data[data['title_field'] == 'Vacant Housing Units'].value)
vacant_houses = vacant_houses[vacant_houses['value'] == max_vacant_houses]
print(vacant_houses)
max_occuped_houses = max(data[data['title_field'] == 'Occupied Housing Units'].value)
occupied_houses = occupied_houses[occupied_houses['value'] == max_occuped_houses]
print(occupied_houses)
```
